```python
lats = []
lons = []
house_IDs = []

for i in range(0, 100):
    lats.append(house_data.lat[i])
    lons.append(house_data.long[i])
    house_IDs.append(house_data.id[i])

house_GPSs = list(zip(lats, lons))
popups = ['house_id: '+'{}'.format(info) for info in house_IDs]
```

```python
from folium.plugins import MarkerCluster
import numpy as np
house_map = folium.Map(location=[np.mean(lats), np.mean(lons)],   
                  tiles='Cartodb Positron', width='80%')
house_map.add_child(MarkerCluster(locations=house_GPSs, popups=popups))
house_map
```
